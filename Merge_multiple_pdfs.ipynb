```python
import fitz  # PyMuPDF
# Open the PDF file
pdf_path = "Excel notes.pdf" #Put your pdf file path in “--”
doc = fitz.open(pdf_path)
# Extract text from each page
for page_num in range(len(doc)):
    page = doc[page_num]
    text = page.get_text()
    print(f"Page {page_num + 1}:\n{text}")
# Close the document
doc.close()
```

```python
from PyPDF2 import PdfMerger
# List of PDF files to merge
pdf_files = ['Excel notes.pdf', 'MP_INTERN_REPORT.pdf', 'Eye Disease Classification using Deep Learning.pdf']
# Create a PDF merger object
merger = PdfMerger()
# Append each file
for pdf in pdf_files:
    merger.append(pdf)
# Write the merged PDF to a file
merger.write("merged.pdf")
merger.close()
print("PDFs merged successfully into 'merged.pdf'")
```
